```python
#define count noise variable which groups complaint types by unique key, counts them and sorts them on a descending order
df3_CountNoise = df2.groupby('Complaint Type')['Unique Key'].count().sort_values(ascending = False)
#add a total variable to sum the number of noise related complaints
total = df3_CountNoise.sum()
print (df3_CountNoise)
#The print(f"\nTotal: {total}") statement prints the value of the total variable on a new line. 
#The f before the string indicates that it is an f-string, which allows you to embed expressions inside string literals,
#using {}. In this case, the expression is total, which is the sum of the values in the Unique Key column grouped by 
#Complaint Type. The \n character is used to print the value on a new line.
"print (f\nTotal: {total}")
```

```python
# Data
complaints = ['Noise - Residential', 'Noise - Street/Sidewalk', 'Noise', 'Noise - Commercial', 'Noise - Vehicle', 'Noise - Other']
counts = [5838, 1866, 1805, 1297, 1290, 167]

# Plot Pie Chart. autopct='%1.1f%%' displays percentage with one decimal
plt.pie(counts, labels=complaints, autopct='%1.1f%%')
plt.title('Complaint Types')
plt.show()
```

```python
df4_CntNoiseBorough = df2.groupby('Complaint Type')['Borough'].count().sort_values(ascending = False)
#add a total variable to sum the number of noise related complaints
total = df4_CntNoiseBorough.sum()
print (df4_CntNoiseBorough)
print (f"\nTotal: {total}")
```

```python
#Breakdown the complaint type by Borough
df5_CntNoiseBorough = df2.groupby(['Complaint Type', 'Borough'])['Unique Key'].count().reset_index(name='Count')
total = df5_CntNoiseBorough['Count'].sum()
print(df5_CntNoiseBorough)
print(f"\nTotal: {total}")
```

```python
pivot = df5_CntNoiseBorough.pivot_table(index =['Complaint Type','Borough'], 
                       values =['Complaint Type'], 
                       aggfunc ='sum') 
print(pivot) 
```
