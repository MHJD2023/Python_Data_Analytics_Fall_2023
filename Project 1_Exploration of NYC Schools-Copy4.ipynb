# Cleaning Tasks

```python
# Segregate District, Borough and School Number from the original School DBN in new columns, so as to help with filtering
df['District'] = df['School DBN'].apply(lambda x: x[:2])
df['Borough'] = df['School DBN'].apply(lambda x: x[2])
df['School Number'] = df['School DBN'].apply(lambda x: x[-3:])
df_DBN = df[['School DBN', 'School Name', 'School Level', 'Regents Exam', 'Year',
              'Percent Scoring 80 or Above', 'District', 'Borough','School Number']]
```
